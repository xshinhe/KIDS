```python
import pykids

plat = pykids.lib.usePlatform('CPU-MPI')
PM = pykids.lib.Param('../build/param_cmsh_nad1d_morse3c.json', 
                      pykids.lib.Param.fromFile)

if 'DS' in dir(): # note we not allow reallocating of DataSet
    del(DS)
DS = pykids.lib.DataSet()

model = pykids.lib.defaultModelFactory("NAD1D")
sys = pykids.lib.System(model, PM, DS)

solver0 = pykids.lib.defaultSamplingFactory("Sampling", sys)
solver1 = pykids.lib.defaultSolverFactory("NaF", sys)

context1 = pykids.lib.Context(plat, sys, [[solver0,   solver1]])
context2 = pykids.lib.Context(plat, sys, [[solver0], [solver1]])

appl1 = pykids.lib.app.Population("CMM", sys)
appl2 = pykids.lib.app.LinearSpectrum("CMM", sys)
appl3 = pykids.lib.app.ReactionRate("CMM", sys, 'side-flux')
context.setApplications([appl1, appl2, appl3])

context.run()     # run first set of solvers
context.run()     # run next set of solvers
# context.runall()  # run all at once

context.summary()
df = context.result().asDataFrame()
```
